```python
import cv2
import sys
import numpy as np
import time
import datetime
import os
import warnings
warnings.filterwarnings("ignore")
from keras.models import load_model
from keras.models import model_from_json
import json
from PIL import Image
from keras.preprocessing import image
import Levenshtein
from collections import OrderedDict
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
import sys
if not os.path.exists("video_shots_cut"):
    os.mkdir("video_shots_cut")
    
start_time = time.time()
minLineLength = 100
maxLineGap = 50
json_file = open('20_model.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)
loaded_model.load_weights("20_model.h5")
```

```python
video = r'C:\Users\datta\Desktop\tfidf\videos\Week 1-Lecture 1 _ Introduction.mp4'
count = 0
flag = 0
start =0
cap = cv2.VideoCapture(video)
fps = cap.get(cv2.CAP_PROP_FPS)
total_frames=int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
s=""
result_dict={}
rrr=-1
shots=0
while True:
    rrr=rrr+1
    ret, img = cap.read()
    if ret == False: # Check if last frame reached
        print("Video Played till end")
        if flag==1:
            print(' to '+str(round(rrr/fps,2)))
            flag=0
            end=(rrr/fps)-1
            shots=shots+1
            if(end-begin>2):
                ffmpeg_extract_subclip(video,begin,end, targetname="video_shots_cut/cut_"+str(shots)+".mp4")
        cap.release()
        cv2.destroyAllWindows()
        break
    slide_count=0
    resized = cv2.resize(img, (64,64), interpolation = cv2.INTER_AREA)
    resized=resized.reshape([-1,64,64,3])
    result = loaded_model.predict(resized)
    if (result[0][0]==1 and flag==0):
        flag=1
        begin=(rrr/fps)+1
        print('Slide from ->'+str(round(rrr/fps,2)),end=" ")
    else:
        if (result[0][0]==0 and flag==1):
            print(' to '+str(round(rrr/fps,2)))
            flag=0
            end=(rrr/fps)-1
            shots=shots+1
            if(end-begin>2):
                ffmpeg_extract_subclip(video,begin,end, targetname="video_shots_cut/cut_"+str(shots)+".mp4")
    if cv2.waitKey(33) & 0xFF == ord('q'): # Check if user pressed q to QUIT
        print('Keyboard interrup, video exiting')
        break
print(result_dict)        
end_time = time.time()
exec_time = round(end_time - start_time,0)
print("Execution time {}".format(str(datetime.timedelta(seconds=exec_time))))
cap.release()
cv2.destroyAllWindows()
```
